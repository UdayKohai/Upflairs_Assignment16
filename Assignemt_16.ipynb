```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data_frame = pd.read_csv('StudentsPerformance.csv')
data_frame
```

```python
##Graph no 1 (scatter plot)
sns.relplot(x='math score',y='reading score',data=data_frame,hue='gender', style='race/ethnicity',col='race/ethnicity')
plt.show()
# by this scatter plot we can see all of the race's relation of math score with reading score is positive 
# also we can we see there are more rocords of Group B and Gruop C
# and least no of records are of group A
```

```python
# Graph 2(line graph)
sns.relplot(x='reading score',y='writing score', data=data_frame,kind='line')
plt.title("Line graph for reading and writing score")
plt.show()
# By this line graph we can see there is positive retaion between reading score and writting score 
# 
```

```python
# Graph 3 (bar graph)
sns.catplot(x='parental level of education',y='math score',data=data_frame,kind='bar',hue='gender')
plt.title('Relation of Parent Education and Math Score')
plt.xlabel('Parent Education')
plt.ylabel('Maths Score')
plt.xticks(rotation=60)
plt.show()
# by this bar graph we can see that 
# for males : 
#       whoes parents have Masters's degree have score highest in maths 
#       and whoes parent have highschool education are scoring the least in maths
# for Females:
#       whose parent have bachlelor's degree  are scoring highest in maths
#       while whose parent have high school or some highschool education are scoring least in maths
```

```python
#Graph 4 (pie chart)
lbl=list(data_frame['race/ethnicity'].value_counts().keys())
val=list(data_frame['race/ethnicity'].value_counts().values)
plt.pie(x=val,labels=lbl,wedgeprops={'ec':'k'},explode=[0.2,0.02,0.02,0.02,0.3],autopct='%1.1f%%')
plt.show()
# By this graph we can see that group C has the most no of students records accounting for (31.9%) of all the records
# while on the other hand group A has the least no of studenr record accounting for only (8.9%) of all the records
```
